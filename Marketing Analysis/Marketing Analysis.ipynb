# Project: Marketing Analysis using Python
## Goal: To visualize the trend of Sales wrt Marketing expenditure

```python
# Importing Relevant Libraries
import matplotlib.pyplot as plt
```

```python
# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [200, 220, 250, 270, 300, 330]
marketing_spend = [50, 60, 65, 70, 90, 120]
```

```python
# Create figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Sales data on primary y-axis
ax1.plot(months, sales, color='blue', marker='o', linestyle='-', linewidth=2, label='Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales (in $1000)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create secondary y-axis for Marketing Spend
ax2 = ax1.twinx()
ax2.plot(months, marketing_spend, color='green', marker='s', linestyle='--', linewidth=2, label='Marketing Spend')
ax2.set_ylabel('Marketing Spend (in $1000)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Title and grid
plt.title('Monthly Sales vs Marketing Spend')
ax1.grid(True, linestyle=':', linewidth=0.5)

# Combine legends from both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```
