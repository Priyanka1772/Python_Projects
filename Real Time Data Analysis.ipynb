# Stock Price Analysis (Financial Data Project)

## Project Goal: Analyze stock trends and moving averages.

```python
# Importing relevant libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample Stock Prices

prices = np.array([100, 102, 101, 105, 107, 110, 108, 109, 115])
```

```python
# Moving average using convolution

window_size = 3
weights = np.ones(window_size) / window_size
moving_avg = np.convolve(prices, weights, mode='valid')
```

```python
# Plotting: Line Chart

plt.plot(prices, label='Price')
plt.plot(range(window_size-1, len(prices)), moving_avg, label='Moving Avg')
plt.legend()
plt.show()
```
