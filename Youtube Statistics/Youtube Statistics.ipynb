## DATASET DESCRIPTION 
 1. Videos-stats
 2. Comments: 18409 rows × 5 columns.

Attributes information:

1. videos-stats.csv:
● Title: Video Title.
● Video ID: The Video Identifier.
● Published At: The date the video was published in YYYY-MM-DD.
● Keyword: The keyword associated with the video.
● Likes: The number of likes the video received. If this value is -1, the likes are not publicly
visible.
● Comments: The number of comments the video has. If this value is -1, the video creator
has disabled comments.
● Views: The number of views the video got.

2. comments.csv:
● Video ID: The Video Identifier.
● Comment: The comment text.
● Likes: The number of likes the comment received.
● Sentiment: The sentiment of the comment. A value of 0 represents a negative sentiment,
while values of 1 or 2 represent neutral and positive sentiments respectively.

# QUERIES


*EXTRACTING YEAR, MONTH AND DAY AS SEPARATE COLUMNS FROM*
***"PUBLISHED_AT" COLUMN TO "PUBLISHED_YEAR", "PUBLISHED_MONTH" AND "PUBLISHED_DAY"***

```python
videos['Published_year']=videos['Published_At'].dt.year
videos['Published_month']=videos['Published_At'].dt.month_name()
videos['Published_day']=videos['Published_At'].dt.day_name()
videos.style.background_gradient(cmap='Purples')
```

# QUERY-1

**QUERY FOR CHECKING THE GROWTH OF CONTENT ON YOUTUBE PER YEAR IN FORM OF A PLOT**

```python
df= videos['Keyword'].groupby(videos['Published_year']).value_counts().unstack('Published_year').sum(axis=0)
# df=df.unstack('Published_year')
# df=df.sum(axis=0)
# df
```

```python
plt.figure(figsize = (15,5))
plt.plot(df, marker='*',color='purple')
plt.xlabel('Year',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.xticks(size = 10, rotation=40, ha="right")
plt.yticks(size = 10)
plt.title('Videos added over the years on Youtube', fontsize=20, fontweight="bold")
plt.grid()
```

**So, this graph represents the growth of content on Youtube per year.**

# QUERY-19

**QUERY FOR PRESENTING VIDEOS ADDED ON yOUTUBE MONTHLY IVER THE YEARS IN THE FORM OF A PLOT**

```python
df=merged[['Published_month','Published_year']]
keyword = df[df['Published_year']>2016]
keyword = (
     keyword
    [keyword['Published_month'].isin(keyword['Published_month'].value_counts().index)]
    .value_counts().to_frame()
    .reset_index()
    .rename(columns={0:'count'})
    .pivot('Published_month', 'Published_year', 'count')
 .fillna(0)
     )
keyword
```

```python
plt.figure(figsize=(18,10))
ax = sns.heatmap(keyword, annot=True, fmt="f", cmap='Purples')
plt.ylabel('Month',fontsize = 15)
plt.xlabel('Year',fontsize = 15)
plt.title('Videos added on youtube monthly over the years',fontsize = 18)
```
